# Reshape data for plotting in ggplot2

```python
import pandas as pd
```

## American Ballet Theatre

```python
df_abt = pd.read_csv('abt-raw.csv')
df_abt.head()
```

```python
df_abt_long = pd.melt(df_abt, id_vars=['name'], value_vars=['Started training', 'Joined a different company',
       'Studied at a prestigious school', 'Studied at ABT JKO school',
       'Promoted to principal at a different company',
       'Joined ABT Studio Company', 'Became an apprentice',
       'Became a corps de ballet member', 'Became a soloist',
       'Became a principal'])

df_abt_long.head()
```

```python
df_abt_long.to_csv('abt.csv', index=False)
```
